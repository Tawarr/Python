```
import re

def is_palindrome(text):

    normalized_text = text.lower()

    cleaned_text = re.sub(r'[^a-z0-9]', '', normalized_text)

    reversed_text = cleaned_text[::-1]

    return cleaned_text == reversed_text

if __name__ == "__main__":
    test_phrases = [
        "racecar",
        "hello world",
        "Madam, I'm Adam", # Phrase with punctuation and capitalization
        "A man, a plan, a canal: Panama",
        "level",
        "Python",
        "noon"
    ]

    print("--- Palindrome Checker Results ---")
    for phrase in test_phrases:
        result = is_palindrome(phrase)

        # Determine the status string for output
        status = "IS a Palindrome" if result else "is NOT a Palindrome"

        print(f"'{phrase}' -> {status}")

    # You can also test with user input
    print("\n--- Custom Input Test ---")
    try:
        user_input = input("Enter a word or phrase to check: ")
        if user_input:
            if is_palindrome(user_input):
                print(f"Yes, '{user_input}' is a Palindrome!")
            else:
                print(f"No, '{user_input}' is NOT a Palindrome.")
        else:
            print("No input provided.")
    except EOFError:
        # Handles environments where input() might fail (like some runtimes)
        pass
```
