```
def is_palindrome(word):
    return word == word[::-1]

def check_word():
    word = input("Enter a word: ")
    word.lower()
    if is_palindrome(word):
        print(word, "is a palindrome")
    else:
        print(word, "is NOT a palindrome")


def find_palindromes_in_sentence():
    sentence = input("Enter a sentence: ")
    sentence.lower()
    words = sentence.split()
    palindromes = [w for w in words if is_palindrome(w)]

    print("Palindromes found:")
    if palindromes:
        for w in palindromes:
            print(" -", w)
    else:
        print("No palindromes found.")


def find_palindromes_in_file():
    try:
        filename = input("Enter file name (with .txt): ")
        with open(filename, "r") as file:
            content = file.read()
            content.lower()

        words = content.split()
        palindromes = [w for w in words if is_palindrome(w)]

        print("\nPalindromes found in file:")
        if palindromes:
            for w in palindromes:
                print(" -", w)
        else:
            print("No palindromes in file.")

    except FileNotFoundError:
        print("File not found. Try again!")


def save_output():
    text = input("\nEnter some text: ")
    text.lower()
    words = text.split()
    palindromes = [w for w in words if is_palindrome(w)]

    with open("palindromes_output.txt", "w") as f:
        f.write("Palindromes found:\n")
        for w in palindromes:
            f.write(w + "\n")

    print("Output saved to 'palindromes_output.txt'")


def menu():
    while True:
        print("\n========== PALINDROME WORDS FINDER ==========")
        print("1. Check a single word")
        print("2. Find palindromes in a sentence")
        print("3. Find palindromes inside a text file")
        print("4. Save palindrome results to a file")
        print("5. Exit")

        choice = input("\nEnter your choice (1-5): ")

        if choice == "1":
            check_single_word()
        elif choice == "2":
            find_palindromes_in_sentence()
        elif choice == "3":
            find_palindromes_in_file()
        elif choice == "4":
            save_output()
        elif choice == "5":
            print("Exiting...Goodbye!")
            break
        else:
            print("Invalid choice. Try again!")

# Start the program
menu()
```
